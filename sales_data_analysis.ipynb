```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Sample categories and products
categories = {
    "Electronics": ["Smartphone", "Laptop", "Headphones", "Smartwatch"],
    "Furniture": ["Chair", "Table", "Sofa", "Bookshelf"],
    "Clothing": ["T-shirt", "Jeans", "Jacket", "Shoes"]
}

# Generate random data
data = []
start_date = datetime(2024, 1, 1)

for order_id in range(1, 51):  # 50 rows
    category = random.choice(list(categories.keys()))
    product = random.choice(categories[category])
    quantity = random.randint(1, 10)
    unit_price = round(random.uniform(10, 1000), 2)
    date = start_date + timedelta(days=random.randint(0, 365))

    data.append([order_id, date.strftime("%Y-%m-%d"), category, product, quantity, unit_price])

# Create DataFrame
df = pd.DataFrame(data, columns=["Order ID", "Date", "Category", "Product", "Quantity", "Unit Price"])

# Save to CSV
df.to_csv("sales_data.csv", index=False)

print("Dataset created: sales_data.csv")
print(df.head())
```

```python
# Calculate total sales
df['Total Sales'] = df['Quantity'] * df['Unit Price'] 
df
```

```python
#total revenue
total_revenue = df['Total Sales'].sum()
print(f"Total Revenue: ${total_revenue:.2f}")

#average unit price
avg_price = df['Unit Price'].mean()
print(f"Average Unit Price: ${avg_price:.2f}")

#Total quantity sold
total_quantity = df['Quantity'].sum()
print(f"Total Quantity Sold: {total_quantity}")
```

```python
#creating sales by category
sales_by_category = df.groupby('Category')['Total Sales'].sum().reset_index()
sales_by_category = sales_by_category.sort_values('Total Sales', ascending=False)
print(sales_by_category)
```

```python
#creating sales by product
sales_by_product = df.groupby('Product')['Total Sales'].sum().reset_index()
sales_by_product = sales_by_product.sort_values('Total Sales', ascending=False)
print(sales_by_product) 
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
df.head()
```

```python
#calculate daily sales
daily_sales = df.groupby('Date')['Total Sales'].sum().reset_index()
daily_sales = daily_sales.sort_values('Date')
print(daily_sales)
```

```python
#calculate monthly sales
month_sales = df.groupby(df['Date'].dt.to_period('M'))['Total Sales'].sum().reset_index()
month_sales['Date'] = month_sales['Date'].dt.to_timestamp()
month_sales['month_name'] = month_sales['Date'].dt.month_name()
month_sales = month_sales[['month_name', 'Total Sales']]
print(month_sales)
```

```python
import matplotlib.pyplot as plt

# Plot daily sales
plt.figure(figsize=(12, 6))
plt.bar(month_sales['month_name'].astype(str), month_sales['Total Sales'])
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.savefig('monthly_sales.png')
```

```python
# Plot sales by category
plt.pie(sales_by_category['Total Sales'], labels=sales_by_category['Category'], autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.figure(figsize=(8, 8))
plt.show()
plt.savefig('sales_by_category.png')

# Plot sales by product
plt.bar(sales_by_product['Product'], sales_by_product['Total Sales'])
plt.title('Sales Distribution by Product')
plt.xlabel('Product')
plt.xticks(rotation=45)
plt.tight_layout()
plt.ylabel('Total Sales ($)')
plt.figure(figsize=(10, 10))
plt.show()
plt.savefig('sales_by_product.png')
```

```python
# Moving average over 3 months
month_sales['3_month_MA'] = month_sales['Total Sales'].rolling(window=3).mean()

# Plot
plt.figure(figsize=(10,6))
plt.plot(month_sales['month_name'] ,month_sales['Total Sales'], label='Monthly Sales', marker='o')
plt.plot(month_sales['3_month_MA'], label='3-Month Moving Average', color='orange')
plt.legend()
plt.title("Sales with Moving Average Trend")
plt.tight_layout()
plt.show()
plt.savefig('sales_with_moving_average.png')
```
